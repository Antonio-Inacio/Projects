```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None) #seta pra mostrar todas as colunas
pd.set_option('display.expand_frame_repr', False) #faz o dataframe n quebrar em varias linhas
```

```python
arquivo = './Datasets/amazon.csv' #caminho do arquivo
```

```python
#leitura do arquivo
df = pd.read_csv(arquivo,sep=",",index_col=None)
df.head(5)
```

```python
def clean_currency(x):
    """
    Se o valor for texto, remove o símbolo e as vírgulas.
    Se já for número, retorna o próprio número.
    """
    if isinstance(x, str):
        # Remove o símbolo de Rúpia (₹) e vírgulas, então converte
        return float(x.replace('₹', '').replace(',', ''))
    # Se já for um número (int ou float), converte para float diretamente
    return float(x)

# Aplica a função às colunas
df['discounted_price'] = df['discounted_price'].apply(clean_currency)
df['actual_price'] = df['actual_price'].apply(clean_currency)
```

```python
# Verificar o resultado
print(df[['discounted_price', 'actual_price']].head(50))
print(df[['discounted_price', 'actual_price']].dtypes)
```

```python
#Salva o arquivo limpo
df.to_csv('./Cleaned_Datasets/amazon_cleaned.csv', index=False)
```
